```python
import pandas as pd
```

```python
customer_info = pd.read_excel("1.1 고객정보.xlsx")
product_info = pd.read_excel("2.1 상품정보.xlsx")
order_info = pd.read_excel("3.1 주문정보.xlsx")
pay_info = pd.read_excel("4.1 결제정보.xlsx")
refund_info = pd.read_excel("5.1 환불정보.xlsx")
```

```python
# 중복하지 않은 상품명 검색
set(order_info['상품명'].unique()) - set(product_info['상품명'].unique()) # 상품정보에 없는 상품명 확인
set(product_info['상품명'].unique()) - set(order_info['상품명'].unique()) # 주문정보에 없는 상품명 확인
```

고객/주문/결제/상품/환불 merge

```python
order_pay_df = pd.merge(order_info, pay_info, how='outer', left_on='주문번호', right_on='주문번호')
order_pay_df.head()
```

```python
order_pay_refund_df = pd.merge(order_pay_df, refund_info, how='outer', left_on='주문번호', right_on='주문번호')
order_pay_refund_df.head()
```

```python
order_pay_refund_product_df = pd.merge(order_pay_refund_df, product_info, how='outer', left_on='상품번호', right_on='상품번호')
order_pay_refund_product_df.head()
```

```python
result = pd.merge(customer_info, order_pay_refund_product_df, how='outer', left_on='회원번호', right_on='회원번호')
result.head()
```

```python
result.to_excel("최종.xlsx")
```

고객/주문/결제 merge

```python
customer_order = pd.merge(customer_info, order_info, how='outer', left_on='회원번호', right_on='회원번호')
```

```python
customer_order_pay = pd.merge(customer_order, pay_info, how='outer', left_on='주문번호', right_on='주문번호')
```

```python
customer_order_pay.to_excel('최종_2.xlsx', index=False)
```
